```python
import numpy as np
np.random.seed(42)
def get_data():
    """
    Generates synthetic data for a 2-feature, 2-output linear regression task using NumPy.
    Outputs are:
        y1 = 2*x1 + 3*x2 + 5 + noise1
        y2 = -x1 + 4*x2 + 2 + noise2
    Returns two outputs:
        X: feature matrix of shape (n_samples, 2), containing x1 and x2
        y: target matrix of shape (n_samples, 2), containing y1 and y2 for each sample
    """

    n_samples = 20

    # Generate random integers for x1 and x2
    x1 = np.random.randint(0, 11, size=(n_samples, 1))
    x2 = np.random.randint(0, 11, size=(n_samples, 1))

    # Stack into feature matrix X
    X = np.hstack((x1, x2))

    # Generate noise for each output
    # noise1 = np.random.uniform(-2, 2, size=(n_samples, 1))
    # noise2 = np.random.uniform(-2, 2, size=(n_samples, 1))
# 
    # Compute targets y1 and y2
    y1 = 2 * x1 + 3 * x2 + 5 #+ noise1
    y2 = -1 * x1 + 4 * x2 + 2 #+ noise2

    # Stack targets into (n_samples, 2)
    y = np.hstack((y1, y2))

    return X, y
```

```python
# X is a list of (x1, x2) pairs, y is a list of (y1, y2) pairs
X, y = get_data() 
print("Shape of X:",X.shape)
print(X[:5])
print("Shape of y:",y.shape)
print(y[:5])
# Refer to this if you don't understand shapes. https://www.geeksforgeeks.org/python/numpy-array-shape/ 
```
